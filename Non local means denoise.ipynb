```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('sample.jpg')
dst = cv.fastNlMeansDenoising(img,None,h=10,templateWindowSize=7,searchWindowSize=21)
plt.subplot(121),plt.imshow(img)
plt.subplot(122),plt.imshow(dst)
plt.show()
```

```python
plt.imsave('denoised.jpg', dst)
```
